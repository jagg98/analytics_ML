## Nombre:

Se tiene una base de datos de evaluación de automóviles, en donde cada automovil se ha caracterizado mediante 6 atributos:  

1. La precio de compra, llamado *buying* , que puede tener los siguientes valores: vhigh, high, med, low. 
2. El costo del mantenimiento, llamado *maint* que puede tener los siguientes valores: vhigh, high, med, low. 
3. El número de puertas, llamado *doors*, que puede tener los siguientes valores: 2, 3, 4, 5more. 
4. La capacidad experesada en el número de personas, llamado *persons*, que puede tener los siguientes valores: 2, 4, more. 
5. El tamaño del portamaletas,llamado *lug_boot*, que puede tener los siguientes valores: small, med, big. 
6. La seguridad estimada del auto, llamado *safety*,que puede tener los siguientes valores: low, med, high. 

De acuerdo con estas seis características, el auto puede ser clasificado en inaceptable (*unacc*), aceptable (*acc*), bueno (*good*) o muy bueno (*vgood*).

Le piden construir un sistema experto que clasifique automáticamente los autos utilizando la base de datos mencionada para su entrenamiento y validación, y el modelo de perceptrón como algoritmo de aprendizaje de máquina. Para la construcción del sistema tenga en cuenta las siguientes condiciones:

1. La base de datos y su descripción se encuentra el la página https://archive.ics.uci.edu/ml/datasets/Car+Evaluation. Debe descargar los datos y separarlos de las etiquetas de forma correcta. 
2. El perceptron no acepta variables categóricas como entradas o salidas (letras), por tanto debe convertir las variables que tiene a valores numéricos (variables numéricas). La forma en se haga la conversión es libre, perodebe ser especificada en los comentarios del código. Recuerde que debe hacer esto tanto para los datos como para las etiquetas.
3. Se debe tener una etapa de preprocesamiento.
4. Se debe mostrar el acierto de clasificación.
5. Se debe mostrar la etiqueta del conjunto de validación. La etiqueta debe estar en los términos inaceptable (*unacc*), aceptable (*acc*), bueno (*good*) o muy bueno (*vgood*).
6. Si requiere alguna información que no está especificada en las condiciones, la debe suponer y justificar correctamente.

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
from numpy import genfromtxt
import neurolab as nl

data = genfromtxt('car.txt', delimiter=',',dtype=str)


X,y = data[:,:-1],data[:,-1]


y2 = np.zeros((y.shape[0],2))

## Procedo a convertir en valores numericos cada caracteristicas
#creo un diccionario con claves que tienen un valor el cual lo asignaré despues.
buying = {'vhigh': 0,'high': 1,'med':2,'low':3}
maint = {'vhigh': 0,'high': 1,'med':2,'low':3}
doors = {'2':0,'3':1,'4':2,'5more':3}
persons ={'2':0,'4':1,'more':2}
lug_boot = {'small':0,'med':1,'big':2}
safety = {'low':0,'med':1,'high':2}
caracteristica = {'unacc':0,'acc':1,'good':2,'vgood':3}


#estos comentarios es para la profe mas tierna de la U :D (prohibido la utilizacioón de los mismos, gracias :) )
#sin más preambulos empezamos
#Como la salida tiene 4 salidas necesitos dos bits para poder reprecentar la salida en la neurona
a = len(y)
for x in range(a):
    if y[x] == 'unacc':
        y2[x,0] = 0
        y2[x,1] = 0
    if y[x] == 'acc':
        y2[x,0] = 0
        y2[x,1] = 1
    if y2[x] == 'good':
        y2[x,0] = 1
        y2[x,1] = 0
    if y[x] == 'vgood':
        y2[x,0] = 1
        y2[x,1] = 1
    

b = len(X)

corrCarac = X[:,1]

for x in range(a):
    #asigno, de acuerdo a una llave, su respectivo valor
    X[x,0] = buying.get(X[x,0])
    X[x,1] = maint.get(X[x,1])
    X[x,2] = doors.get(X[x,2])
    X[x,3] = persons.get(X[x,3])
    X[x,4] = lug_boot.get(X[x,4])
    X[x,5] = safety.get(X[x,5])

X = X.astype(np.float)
#y = y.astype(np.int64)
# y = np.vstack(y)


X_train, X_test, y_train, y_test = train_test_split(X,y2, test_size = 0.2, random_state = 5)

```

```python
#Aquí relalizo el preprocesamiento de los datos
def remocion(X_train, X_test):
    media = X_train.mean(axis = 0)
    desviacion = X_train.std(axis = 0)
    X_train_prepro = (X_train - media)/desviacion
    X_test_prepro = (X_test - media)/desviacion
    return X_train_prepro, X_test_prepro


X_train_preproR, X_test_preproR = remocion(X_train, X_test)

```

```python
#Profe aquí coloco la dimenciones max y min de los vectores de caracteristicas
dim1min,dim1max = X_train_preproR[:,0].min(), X_train_preproR[:,0].max()
dim2min,dim2max = X_train_preproR[:,1].min(), X_train_preproR[:,1].max()
dim3min,dim3max = X_train_preproR[:,2].min(), X_train_preproR[:,2].max()
dim4min,dim4max = X_train_preproR[:,3].min(), X_train_preproR[:,3].max()
dim5min,dim5max = X_train_preproR[:,4].min(), X_train_preproR[:,4].max()
dim6min,dim6max = X_train_preproR[:,5].min(), X_train_preproR[:,5].max()

print(dim1min,dim1max)
```

```python
#defino mi neurona
nn = nl.net.newff([[dim1min, dim1max],
                   [dim2min, dim2max],
                   [dim3min, dim3max],
                   [dim4min, dim4max],
                   [dim5min, dim5max],
                   [dim6min, dim6max]], [10,5, 2]) #es importante dos capas, baja el error 
nn.train(X_train_preproR, y_train, epochs = 200, show =10, goal = 0.01)
```

```python
output = nn.sim(X_test_preproR)
print(output)
```

```python
#en esta parte combierto mi matriz en 0 y 1 de acuerdo a un umbral para luego compararla con mi y_test
output2 = np.zeros(output.shape)
umbral = 0.8
for i in range(output.shape[0]):
    for j in range(output.shape[1]):
        if output[i,j] < umbral:
            output2[i,j] = 0
        if output[i,j] >= umbral:
            output2[i,j] = 1
print(output2)
error = 0

#Proceso de comparación para mirar el error de mi neurona
for i in range(output.shape[0]):
    print(i,output2[i,0],y_test[i,0])
    print(i,output2[i,1],y_test[i,1])
    if output2[i,0] != y_test[i,0] or output2[i,1] != y_test[i,1]:
        error = error + 1
print((error*100)/output.shape[0],'% de error')
output3 = np.zeros(y_test.shape[0])
output3 = output3.astype(str)

#Proceso de mostrar mi y_test en terminos de los nombres de las etiquetas

for h in range(y_test.shape[0]): 
    a = y_test[h,0]
    b = y_test[h,1]
    if a == 0 and b == 0:
        output3[h] = 'unacc'
    if a == 0 and b == 1:
        output3[h] = 'acc'
    if a == 1 and b == 0:
        output3[h] = 'good'
    if a == 1 and b == 1:
        output3[h] = 'vgood'
print(np.vstack(output3))

#aquí acaba mi codigo, muchas gracias por sintonizarme nos vemos en la próxima.
```
